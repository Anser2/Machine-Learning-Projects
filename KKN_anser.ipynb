```
from scipy import misc
import cv2
from math import sqrt
from numpy import ndarray



for i in range(50):
    G = im[:,:,:,i]
    G = np.reshape(G,[-1], order="F")

    #Read your image here
####### Your code here #######
my_img = cv2.imread(test_image)
my_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
my_img.shape


############################


    #Resize your image
####### Your code here #######
my_img_resized = cv2.resize(my_img, (32, 32))
my_img_resized.shape
plt.imshow(my_img_resized)


##############################


    #Reshape your image as we reshape the image of dataset
####### Your code here #######

my_img_flattened = np.reshape(my_img_resized, [-1], order = "F")
my_img_flattened.shape

#############################
    #Calculate Eucleadian distance between your image and dataset
####### Your code here #######
euc_dist = np.zeros(50)
for i in range(0,50):
  euc_dist[i] = np.sqrt(np.sum((G[i] - my_img_flattened)**2))
  print(f"euc_dist from image {i} is: {euc_dist[i]}")

##############################
```

In this Block you need to fill the code for 1-NN.

```
# Write code for 1 NN
#Find min distance
#Fine at which point min value exists
####### Your code here #######
def one_NN(x):
  min_distance = np.min(x)
  min_distance_id = np.argmin(x)
  return min_distance, min_distance_id

x,y = one_NN(euc_dist)
print(f"index of min distance is: {y}")
print(f"min distance is: {x}")
plt.imshow(im[:,:,:,y])
print("Anser you match closest with: ", label[y][0][0])

##############################
```


Write code for 3 NN


```
#3 NN
#Write code for 3 NN
#Find 3 min distances
#Find their instances
#hint: Use for Loop

####### Your code here #######

def three_NN(x):
  min_distance = np.sort(x)[0:3]
  min_distance_id = np.argsort(x)[0:3]
  return min_distance, min_distance_id
a, b = three_NN(euc_dist)
print(f"Anser you match closest with {label[b[0]][0][0]}, and {label[b[1]][0][0]}, {label[b[2]][0][0]}")
for i in range(0,3):
  plt.figure()
  plt.imshow(im[:,:,:,b[i]])
  plt.show()


##############################
```
